# Conditions in Python

    Conditions help the computer think logically

    A condition is used to make decisions in a program.

    Real-life example:

        If it is raining → take umbrella
        Else → go normally

## Simple if Statement:
    Used when you want to run code only if condition is true

    Syntax :
        if condition:
            statement
            
    If condition is false → nothing runs

```python
# Example :
age = int(input("Enter your age: "))

if age >= 18: # age >= 18 -> 12 >= 18
    print("You are eligible to vote")
```

## if–else Statement

    Used when you want to choose between two options
    
    Syntax:

        if condition:
            statement
        else:
            statement


```python
# age = 26
age = int(input("Enter your age: "))
if age >= 18:
    print("You can vote")
else:
    print("You are not eligible to vote")
```

## if–elif–else (Multiple Conditions)

    Used when you have more than two conditions
    
    Syntax:

        if condition1:
            statement
            
        elif condition2:
            statement
            
        elif condition3:
            statement
            
        else:
            statement

```python
# marks = 0-40 = Fail
# marks = 41-60 = C Grade
# marks = 61 - 80 = B grade
# marks = 80 and above  = A Grade

marks = int(input("Enter the  marks :"))

if marks >= 80:
    print("A grade")
elif marks >= 61:
    print("B Grade")
elif marks >= 41:
    print("C grade")
elif marks >= 0:
    print("Fail")
else:
    print("Invalid marks")
```

```python
# marks = 0-40 = Fail
# marks = 41-60 = C Grade
# marks = 61 - 80 = B grade
# marks = 80 - 100  = A Grade
# marks = 30
marks = int(input("Enter your marks: "))
if 80 <= marks <= 100: # X
    print("Grade A") # X
elif 61 <= marks <= 79:   
    print("Grade B") # X
elif 41 <= marks <= 60:
    print("Grade C") # X
elif 0 <= marks <= 40:
    print("Fail")
else:
    print("Enter valid marks")
```

## Nested if (if inside if)

    Used when one condition depends on another condition
    
    Syntax:

        if condition1:
                statement
                
                if condition2:
                    statement
                else:
                    statement
            
        else:
             statement
             
             if condition2:    
               statement
            else:
                statement

```python
# Check age ≥ 18 and has ID card, then allow entry.
age = int(input("Enter the age: "))
if age>=18:
    print("Correct, Next Question")
    
    id_card = input("Do you have ID card? (yes/no)")
    if id_card.lower() == "yes":
        print("Entry allowed")
    elif id_card.lower() == "no":
        print("Entry not allowed")
    else:
        print("Please Enter yes or no only")

else:
    print("Entry Not allowed")
```

```python
# Leap Year
# rules :
# divisible by 4
# (but not divisible by 100) or (except divisible by 400)
# eg : 1900

year = int(input("Enter the year : "))

if year%400==0 or (year%4==0 and year%100!=0):
    print("Leap Year.")
else:
    print("Not a Leap Year.")
#  1900 = not a leap year.
# 2024  = Leap  Year
```

## Logical Conditions (and, or, not)

    Used to combine multiple conditions
    
### *and* Example
    
    (Both conditions must be true)

```python
age = 23
citizen = "Aus"
#    False or True = True
if age >= 18 or citizen == 'Indian':
    print("You can vote")
else:
    print("You can not vote")
```

### *or* Example

    (Any one condition must be true)

```python
age = 12
citizen = "Indian"

if age >= 18 or citizen == 'Indian':
    print("You can vote")
else:
    print("You can not vote")
```

### *not* Example

    (Reverses the condition)

```python
raining = True

if not raining: # if "not False" = True
    print("Go Inside")
else:
    print("Go Outside")
```

### Shorthand if (One-liner if)

    Used when only one statement is needed

```python
age = 13

if age >= 18: print("Eligible")
```

### Shorthand if-else :

```python
age = 22

print("Eligible" if age>=18 else "Not Eligible")
```

## Ternary Operator :
    The ternary operator is a one-line conditional expression.
    It lets you choose between two values based on a condition.
    
    Syntax:

        value_if_true if condition else value_if_false


```python
principle = int(input("Enter the Principle value :"))
rate = int(input("Enter the rate: "))
time = int(input("Enter the Time: "))
simple_interst = (principle * rate * time)/100
print(simple_interst)
```
