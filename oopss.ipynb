### Method Overriding (Runtime Polymorphism)
    When child class overrides parent class method.
    Same method name → different behavior.

    That is runtime polymorphism.

```python
class Animal:
    def speak(self):
        print("Animal makes sound")

class Dog(Animal):
    def speak(self):
        print("Dog barks")

class Cat(Animal):
    def speak(self):
        print("Cat meows")

a = Animal()
d = Dog()
c = Cat()

a.speak()
d.speak()
c.speak()
```

    Python uses the C3 Linearization algorithm to decide this order.

```python
# simple inheritance
class A:
    def show(self):
        print("Class A")

class B(A):
    pass

obj = B()
obj.show()
```

### C3 Linearization Algorithm
    C3 Linearization Algorithm is the method Python uses to decide the Method Resolution Order (MRO) in multiple inheritance.
    In Simple :
                C3 Linearization creates a consistent order in which Python searches classes for methods.

    It ensures three things:
                            Child class comes before parent classes
                            
                            Order of parents in the class definition is preserved
                            
                            A class appears only once in the MRO

```python
class A:
    pass

class B(A):
    pass

class C(A):
    pass

class D(B, C):
    pass

print(D.mro())
```

```python
help(D)
```

#### Why C3 Linearization is Needed
    When there are multiple parent classes, Python must decide which class to search first.

      A
     / \
    B   C
     \ /
      D

    class D(B, C) : Both B and C inherit from A, so Python must avoid searching A twice and must keep the correct order.
                    C3 Linearization resolves this.

##### Encapsulation + Getter & Setter
    Agar private variable hai, to usee access karne ka official tarika hota hai:
    getter → value read karo
    setter → value update karo (rules laga kar)

```python
class Student:
    def __init__(self):
        self.__marks = 0

    def get_marks(self):       # getter
        return f"Marks : {self.__marks}"

    def set_marks(self, value):  # setter
        if 0 <= value <= 100:
            self.__marks = value
        else:
            print("Invalid Marks")

s = Student()
print(s.get_marks()) # read 
s.set_marks(50) # update
print(s.get_marks())
```

##### __marks private hai → direct access nahi
##### set_marks() me rules diye (0–100 valid)
##### Yehi encapsulation + data protection hota hai

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance   # private variable

    def get_balance(self):
        return self.__balance

    def deposit(self, amount): # 15000
        if amount > 0:
            self.__balance += amount # self.__balance = self.__balance + amount
            print(f"Deposited ₹{amount}")
        else:
            print("Deposit amount must be positive!")

    def withdraw(self, amount): # 15000 ya kam
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ₹{amount}")
        else:
            print("Insufficient balance!")


# TESTING
acc = BankAccount(5000)
print("Initial Balance:", acc.get_balance())
acc.deposit(15000)
print("Initial Balance:", acc.get_balance())
acc.withdraw(10000)
print("Initial Balance:", acc.get_balance())
acc.withdraw(600)
print("Initial Balance:", acc.get_balance())
```

## ABSTRACTION
    Abstraction : ander ki functionallity chupana, aur sirf jo show karwana chahiye wahi karwana
    Encapsulation : controlled access, koi aur kuch dekh hee ni skta
##### zaroori cheezon ko dikhana, details chhupaana.
#### Abstraction vs Encapsulation
##### Encapsulation = data hiding & controlled access (private vars, getters/setters).
##### Abstraction = interface hiding complexity (what operations are allowed — not how they are implemented).
#### Abstract classes & methods in Python
### ABC : 
    An abstract class is a class that cannot be used to create objects directly.
    It is used as a blueprint for other classes.
    It ensures that child classes implement certain methods.
    
    Example:
            If we create a class Vehicle, we may want every vehicle to have a method called start().
            
            So we make start() abstract, forcing every child class to implement it.
### What is an Abstract Method?

    An abstract method is a method that:
                                        is declared in the parent class
                                        has no implementation
                                        must be implemented in the child class
    

```python
# import abc
# dir(abc)
```

```python
from abc import ABC, abstractmethod

class Animal(ABC): # Animal = abstract base class
    @abstractmethod
    def sound(self):
        pass
        
class Dog(Animal):
    def sound(self):
        print("Dog barks")

class Cat(Animal):
    def sound(self):
        print("Cat meows")

d = Dog()
d.sound()

c = Cat()
c.sound()
```

```python
## Simple Example — class Shape
# Hum ek Shape abstract class banayenge jiska  abstract method area() har shape (circle, rectangle, square) me alag hoga. 
# Shape me describe() jaisa ek concrete method bhi rakhte hain jo sab shapes ke liye common text de sakta hai.

from abc import ABC, abstractmethod
import math

class Shape(ABC):
    """Abstract base class representing a geometric shape."""

    @abstractmethod
    def area(self):
        """Return area of the shape. Must be implemented by subclasses."""
        pass

    def describe(self): # concrete method
        """Concrete method — available for all subclasses."""
        print("I am a shape. I can calculate my area.")

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius * self.radius

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height
        
class Square(Shape):
    def __init__ (self,lengthh):
        self.lengthh = lengthh

    def area(self):
        return self.lengthh * self.lengthh
# Usage
c = Circle(3)
r = Rectangle(4, 5)
s = Square(7)
c.describe()
print("Circle area:", c.area())

r.describe()
print("Rectangle area:", r.area())

s.describe()
print("Sqaure area:", s.area())
```

```python
# Create an abstract class Notification with:
# prepare_message()
# send_message() (abstract)
# Make:
# WhatsAppNotification
# SlackNotification
# EmailNotification
# Add features like encryption in some classes.

from abc import ABC, abstractmethod

# Simple Caesar cipher for demo
def encrypt(text, shift=3):
    encrypted = ""
    for char in text:
        encrypted += chr(ord(char) + shift)
    return encrypted


# Abstract Base Class
class Notification(ABC):
    
    def __init__(self, message):
        self.message = message
        self.prepared_message = None

    def prepare_message(self):
        # common formatting logic
        self.prepared_message = f"***NOTIFICATION***\n{self.message}"
        print("Message prepared:", self.prepared_message)

    @abstractmethod
    def send_message(self):
        pass


# WhatsApp Notification
class WhatsAppNotification(Notification):
    
    def send_message(self):
        encrypted_msg = encrypt(self.prepared_message)
        print("Sending WhatsApp message (encrypted):")
        print(encrypted_msg)


# Slack Notification
class SlackNotification(Notification):
    
    def send_message(self):
        print("Sending Slack message (no encryption):")
        print(self.prepared_message)


# Email Notification
class EmailNotification(Notification):
    
    def send_message(self):
        encrypted_msg = encrypt(self.prepared_message, shift=5)
        print("Sending Email message (heavier encryption):")
        print(encrypted_msg)



# TESTING
wh = WhatsAppNotification("Meeting at 5 PM.")
wh.prepare_message()
wh.send_message()

print("\n----------------------\n")

sl = SlackNotification("Build completed successfully.")
sl.prepare_message()
sl.send_message()

print("\n----------------------\n")

em = EmailNotification("Your OTP is 4521.")
em.prepare_message()
em.send_message()
```

```python
# Can an abstract class contain no abstract methods? -Yes
from abc import ABC

class Base(ABC):
    def common(self):
        print("Common Method")

obj = Base()
obj.common()
```

```python
# Can we inherit from multiple abstract classes? - Yes

from abc import ABC, abstractmethod

class A(ABC):
    @abstractmethod
    def method_A(self):
        pass

class B(ABC):
    @abstractmethod
    def method_B(self):
        pass

class C(A,B):
    def method_A(self):
        print("A")
    def method_B(self):
        print("B")
    
abc = C()
abc.method_A()
abc.method_B()
```

```python
# What happens if abstract methods are private (__method)?

from abc import ABC, abstractmethod

class A(ABC):
    @abstractmethod
    def __method_a(self): # private method
        pass

class B(A):
    def _A__method_a(self):
        print("Hello")

obj = B()
obj._A__method_a()
```

```python
# Can static methods be abstract? - Yes

from abc import ABC, abstractmethod

class A(ABC):
    @staticmethod
    @abstractmethod
    def method_a(self): # private method
        pass

class B(A):
    @staticmethod
    def method_a():
        print("Hello")

mno = B()
mno.method_a()
```

```python
# What is the difference between NotImplementedError and @abstractmethod?

class A:
    def show(self):
        raise NotImplementedError

abc = A()
# abc.show()
```

```python
from abc import ABC, abstractmethod

class A(ABC):
    @abstractmethod
    def show(self):
        pass

abc = A()
# abc.show()
```

```python
# Create an abstract class MachineLearningModel with abstract methods train() and predict(). Create child classes LinearRegression and RandomForest.

from abc import ABC,abstractmethod

class MachineLearningModel(ABC):
    @abstractmethod
    def train(self):
        pass
    @abstractmethod
    def predict(self):
        pass

class LinearRegression(MachineLearningModel):
    def train(self):
        print("Training Linear Regression Model")

    def predict(self):
        print("Predicting Linear Regression Model")

class RandomForest(MachineLearningModel):
    def train(self):
        print("Training Random Forest Model")

    def predict(self):
        print("Predicting Random Forest Model")

l = LinearRegression()
l.train()
l.predict()

r = RandomForest()
r.train()
r.predict()
```
